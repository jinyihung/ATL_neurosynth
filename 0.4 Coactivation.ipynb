# Co-activation analysis

2. contrasting the co-activation patterns between sets of related regions. That is, we seek to find voxels throught the brain that co-activate to a greater extent with one cluster than others

```python
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
from neurosynth.base.dataset import Dataset
dataset = Dataset.load("data/neurosynth_0.6_400_4.pkl")
```

The following two functions are all that is necessary to perform this analysis:

`coactivation_contrast` peforms the analysis and returns an image for each region in the analysis. For each region, voxels in the image are those that co-activated to a greater extent with that region, than other regions in the analysis.

`make_thresholded_slices` takes a list of images and overlays them using nilearn's plot_roi

```python
from coactivation import coactivation_contrast
from plotting import make_thresholded_slices
```

## N=4

```python
contrast_mas_60w4 = coactivation_contrast(dataset,'images/wATL/whole60_k4.nii', q=0.01)
```

```python
colors = ['magenta','dark green','orange',"dark grey"]
colors = sns.xkcd_palette(colors)
make_thresholded_slices(contrast_mas_60w4, colors, cut_coords=range(-60, 60, 12), display_mode='x', annotate=True)
make_thresholded_slices(contrast_mas_60w4, colors, cut_coords=range(-60, 60, 12), display_mode='y', annotate=True)
make_thresholded_slices(contrast_mas_60w4, colors, cut_coords=range(-60, 60, 12), display_mode='z', annotate=True)
```
